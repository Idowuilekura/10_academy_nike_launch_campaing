```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('dark')
import pandas as pd
pd.set_option('display.max_colwidth', None)
```

```
africa_influencers_df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/africa_influencers_df.csv')
```

```
africa_leaders_df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/africa_leaders_df.csv')
```

```
africa_leaders_df.head()
```

```
top_10_most_reached_leader = africa_leaders_df.sort_values(by='reach_score',ascending=False).head(10)
top_10_most_reached_leader['reach_rank_score'] = top_10_most_reached_leader['reach_score'].rank(ascending=False)
top_10_most_reached_influencer = africa_influencers_df.sort_values(by='reach_score',ascending=False).head(10)
top_10_most_reached_influencer['reach_rank_score'] = top_10_most_reached_influencer['reach_score'].rank(ascending=False)
```

```
fig,(ax1,ax2) = plt.subplots(1,2,figsize=(20,8))
sns.barplot(data=top_10_most_reached_leader,x='twitter_name',y='reach_score',ax=ax1)
ax1.set_yscale('log')
ax1.set_xticklabels(top_10_most_reached_leader['twitter_name'],rotation=75)
ax1.set_title('Top 10 Africa leaders most reached on Twitter(followers - following)')
ax1.set_ylabel('Numbers of Reach')
sns.barplot(data=top_10_most_reached_leader,x='twitter_name',y='reach_rank_score',
ax=ax2)
ax2.set_xticklabels(top_10_most_reached_leader['twitter_name'],rotation=75)
ax2.set_title('Top 10 Africa leaders most reached by rank(1-10) on twitter')
ax2.set_ylabel('Rank degree')
plt.savefig('leaders_reach.png')

```

```
fig,(ax1,ax2) = plt.subplots(1,2,figsize=(20,8))
sns.barplot(data=top_10_most_reached_influencer,x='twitter_names',y='reach_score',ax=ax1)
ax1.set_yscale('log')
ax1.set_xticklabels(top_10_most_reached_influencer['twitter_names'],rotation=75)
ax1.set_title('Top 10 Africa Influencers most reached on Twitter(followers - following)')
ax1.set_ylabel('Numbers of Reach')
sns.barplot(data=top_10_most_reached_influencer,x='twitter_names',y='reach_rank_score',
ax=ax2)
ax2.set_xticklabels(top_10_most_reached_influencer['twitter_names'],rotation=75)
ax2.set_title('Top 10 Africa influencers most reached by rank(1-10) on twitter')
ax2.set_ylabel('Rank degree')
plt.savefig('Influencers_reach.png')
```

```
top_10_most_rel_leaders = africa_leaders_df.sort_values(by='relevance_score',ascending=False).head(10)
```

```
top_10_most_rel_leaders['relevance_score_rank'] = top_10_most_rel_leaders['relevance_score'].rank(ascending=False)
```

```
top_10_most_rel_infl = africa_influencers_df.sort_values(by='relevance_score',ascending=False).head(10)
top_10_most_rel_infl['relevance_score_rank'] = top_10_most_rel_infl['relevance_score'].rank(ascending=False)
```

```
fig,(ax1,ax2) = plt.subplots(1,2,figsize=(20,8))
sns.barplot(data=top_10_most_rel_leaders,x='twitter_name',y='relevance_score',ax=ax1)
ax1.set_yscale('log')
ax1.set_xticklabels(top_10_most_rel_leaders['twitter_name'],rotation=75)
ax1.set_title('Top 10 most relevant Africa leaders on Twitter(@mentions in others tweets)')
ax1.set_ylabel('Relevance numbers')
sns.barplot(data=top_10_most_rel_leaders,x='twitter_name',y='relevance_score_rank',
ax=ax2)
ax2.set_xticklabels(top_10_most_rel_leaders['twitter_name'],rotation=75)
ax2.set_title('Top 10 most relevant Africa leaders by rank(1-10) on twitter')
ax2.set_ylabel('relevance Rank degree')
plt.savefig('leaders_relevance.png')
```

```
fig,(ax1,ax2) = plt.subplots(1,2,figsize=(20,8))
sns.barplot(data=top_10_most_rel_infl,x='twitter_names',y='relevance_score',ax=ax1)
ax1.set_yscale('log')
ax1.set_xticklabels(top_10_most_rel_infl['twitter_names'],rotation=75)
ax1.set_title('Top 10 most relevant Africa Influencers on Twitter(@mentions in others tweets)')
ax1.set_ylabel('Relevance numbers')
sns.barplot(data=top_10_most_rel_infl,x='twitter_names',y='relevance_score_rank',
ax=ax2)
ax2.set_xticklabels(top_10_most_rel_infl['twitter_names'],rotation=75)
ax2.set_title('Top 10 most relevant Africa influencers by rank(1-10) on twitter')
ax2.set_ylabel('relevance Rank degree')
plt.savefig('influencers_relevance.png')
```

```
from wordcloud import WordCloud,STOPWORDS
import matplotlib.pyplot as plt
text = ten_most_popular_africa_influencers.twitter_nam.values
wordcloud = WordCloud(width=3000,height=2000,background_color='black',stopwords=STOPWORDS).generate(str(text))
fig = plt.figure(figsize=(30,20),facecolor='k',edgecolor='k')
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
plt.tight_layout(pad=0)
plt.savefig('africa_influencers_cloud.png')
```
